```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.linear_model import LinearRegression
# !pip install pandas
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
df = fetch_california_housing()
```

```python
dataset=pd.DataFrame(df.data)
```

```python
dataset.shape
```

```python
dataset.info
```

```python
dataset.columns=df['feature_names']
```

```python
dataset.head
```

```python
dataset["MedHouseVal"]=df.target
```

```python
dataset.head
```

```python
x=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import cross_val_score

model1=LinearRegression()
mse=cross_val_score(model1,x,y,scoring='neg_mean_squared_error',cv=5)
mse
mean_mse=np.mean(mse)
mean_mse
```

```python
# Ridge Regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge=Ridge()
params={
    'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,55,60,75,100]
}

ridge_tunning=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)
ridge_tunning.fit(x,y)
```

```python
print(ridge_tunning.best_params_)
print(ridge_tunning.best_score_)
```

```python
# lasso
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

lasso=Lasso()
params={
    'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,45,60,100]
}

mse=cross_val_score(lasso,x,y,scoring='neg_mean_squared_error',cv=5)
mse
mean_mse=np.mean(mse)
mean_mse
lasso_tunning=GridSearchCV(lasso,params,scoring='neg_mean_squared_error',cv=5)
lasso_tunning.fit(x,y)

print(lasso_tunning.best_params_)
print(lasso_tunning.best_score_)

```

```python
# usingtrain and test spliting
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(
    x,y,test_size=0.3,random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python

model1=LinearRegression()
mse=cross_val_score(model1,x_train_scaled,y_train,scoring='neg_mean_squared_error',cv=5)
mse
mean_mse=np.mean(mse)
mean_mse
```

```python
ridge=Ridge()
params={
    'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,55,60,75,100]
}

ridge_tunning=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)
ridge_tunning.fit(x_train_scaled,y_train)
ridge_pred=ridge_tunning.predict(x_test_scaled)
```

```python
print(ridge_tunning.best_params_)
print(ridge_tunning.best_score_)
```

```python
lasso=Lasso()
params={
    'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,45,60,100]
}

mse=cross_val_score(lasso,x_train_scaled,y_train,scoring='neg_mean_squared_error',cv=5)
mse
mean_mse=np.mean(mse)
mean_mse
lasso_tunning=GridSearchCV(lasso,params,scoring='neg_mean_squared_error',cv=5)
lasso_tunning.fit(x_train_scaled,y_train)
y_pred=lasso_tunning.predict(x_test_scaled)

print(y_pred)

print(lasso_tunning.best_params_)
print(lasso_tunning.best_score_)
```

```python
from sklearn.metrics import r2_score

r2_sc=r2_score(y_test,y_pred)
print("lasso:",r2_sc)

r2_ridge=r2_score(y_test,ridge_pred)
print("ridge:",r2_ridge)
```
